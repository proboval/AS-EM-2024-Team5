# Hypothesis 1: Comparing ACO, RSA, and GWO

For Sammon's stress function and Kruskal's stress function, as well as for class level metrics and method level metrics, we will compare the results of RSA, ACO, and GWO using the Kruskal-Wallis test at each k.

```python
class_k_max = class_metrics_results_['k'].max()
method_k_max = method_metrics_results_['k'].max()
class_k_max
```

```python
from scipy.stats import kruskal


def wallis_kruskal_test(df, df_, error_fun, k):
    rsa = df[df['k'] == k]['best_fitness_rsa_' + error_fun].dropna()
    gwo = df[df['k'] == k]['best_fitness_gwo_' + error_fun].dropna()
    aco = df[df['k'] == k]['best_fitness_aco_' + error_fun].dropna()
    ga = df_[df_['k'] == k]['best_fitness_ga_' + error_fun].dropna()
    pso = df_[df_['k'] == k]['best_fitness_pso_' + error_fun].dropna()

    return [kruskal(rsa, gwo, aco, pso, ga).pvalue, aco.mean(), gwo.mean(), rsa.mean(), ga.mean(), pso.mean()]

wallis_kruskal_test(class_metrics_results_, class_metrics_results, 'sammon', 1)
# сделать таблицу ещё одну с данными предыдущей команды
# для методов тоже
```

```python
# Пример кода с исправлением
class_level_hyp1_results = pd.DataFrame(columns=['k', 'sammon',
                                                 'aco_mean_sammon',
                                                 'gwo_mean_sammon',
                                                 'rsa_mean_sammon',
                                                 'ga_mean_sammon',
                                                 'pso_mean_sammon',
                                                 'kruskal',
                                                 'aco_mean_kruskal',
                                                 'gwo_mean_kruskal',
                                                 'rsa_mean_kruskal',
                                                 'ga_mean_kruskal',
                                                 'pso_mean_kruskal',])

for k in range(1, class_k_max + 1, 2):
    print(k)
    res = wallis_kruskal_test(class_metrics_results_, class_metrics_results, 'sammon', k)
    sammon = res[0]
    aco_mean_sammon = res[1]
    gwo_mean_sammon = res[2]
    rsa_mean_sammon = res[3]
    ga_mean_sammon = res[4]
    pso_mean_sammon = res[5]

    res = wallis_kruskal_test(class_metrics_results_, class_metrics_results, 'kruskal', k)
    kruskal_ = res[0]
    aco_mean_kruskal = res[1]
    gwo_mean_kruskal = res[2]
    rsa_mean_kruskal = res[3]
    ga_mean_kruskal = res[4]
    pso_mean_kruskal = res[5]

    class_level_hyp1_results = pd.concat([
        class_level_hyp1_results,
        pd.DataFrame({
            'k': [k],
            'sammon': sammon,
            'aco_mean_sammon': aco_mean_sammon,
            'gwo_mean_sammon': gwo_mean_sammon,
            'rsa_mean_sammon': rsa_mean_sammon,
            'ga_mean_sammon': ga_mean_sammon,
            'pso_mean_sammon': pso_mean_sammon,
            'kruskal': kruskal_,
            'aco_mean_kruskal': aco_mean_kruskal,
            'gwo_mean_kruskal': gwo_mean_kruskal,
            'rsa_mean_kruskal': rsa_mean_kruskal,
            'ga_mean_kruskal': ga_mean_kruskal,
            'pso_mean_kruskal': pso_mean_kruskal,
        })
    ], ignore_index=True)


class_level_hyp1_results['k'] = class_level_hyp1_results['k'].astype(int)
class_level_hyp1_results
```

# Hypothesis 2: Comparing Sammon's error function for train and validation sets using found optimal subset of metrics

```python
class_level_validation_results = pd.read_csv('class_level_validation_results.csv')
method_level_validation_results = pd.read_csv('method_level_validation_results.csv')
```

```python
class_level_validation_results.head()
```

```python
method_level_validation_results.head()
```

According to the hypothesis, when applying the optimal subsets derived from the training set, Sammon's error on the validation set should be lower than 0.1. To test this hypothesis, we employed a one-tailed t-test, where the null hypothesis (H0) was that the mean Sammon's error using the optimal subsets is greater than or equal to 0.1, while the alternative hypothesis (H1) was that the mean error is less than 0.1.

```python
import numpy as np

class_level_validation_results = class_level_validation_results.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
from scipy.stats import ttest_1samp

class_level_hyp2_results = pd.DataFrame(columns=['sammon_p_value', 'kruskal_p_value', 'sammon_mean', 'kruskal_mean'])
test_sammon = ttest_1samp(class_level_validation_results['sammon'], 0.5, alternative='less')
test_kruskal = ttest_1samp(class_level_validation_results['kruskal'], 0.1, alternative='less')

new_row = pd.DataFrame([{
    'sammon_p_value': test_sammon.pvalue,
    'kruskal_p_value': test_kruskal.pvalue,
    'sammon_mean': class_level_validation_results['sammon'].mean(),
    'kruskal_mean': class_level_validation_results['kruskal'].mean()
}])

class_level_hyp2_results = pd.concat([class_level_hyp2_results, new_row], ignore_index=True)

class_level_hyp2_results
```

```python
method_level_hyp2_results = pd.DataFrame(columns=['sammon_p_value', 'kruskal_p_value', 'sammon_mean', 'kruskal_mean'])
test_sammon = ttest_1samp(method_level_validation_results['sammon'].dropna(), 0.05, alternative='less')
test_kruskal = ttest_1samp(method_level_validation_results['kruskal'].dropna(), 0.1, alternative='less')

new_row = pd.DataFrame([{
    'sammon_p_value': test_sammon.pvalue,
    'kruskal_p_value': test_kruskal.pvalue,
    'sammon_mean': class_level_validation_results['sammon'].mean(),
    'kruskal_mean': class_level_validation_results['kruskal'].mean()
}])

method_level_hyp2_results = pd.concat([method_level_hyp2_results, new_row], ignore_index=True)

method_level_hyp2_results
```

```python
def scientific_notation_to_power_of_ten(number):
    number = str(number)
    if 'e' in number:
        base, exponent = number.split('e')
        return f'{float(base):.2f} \\times 10^{{{int(exponent)}}}'
    else:
        return number
```
